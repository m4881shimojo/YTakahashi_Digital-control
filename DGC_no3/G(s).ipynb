```python
import numpy as np
from control import matlab
import matplotlib.pyplot as plt
import numpy.linalg as LA
```

```python
tsample=0.001
T=0.1;alfa=10;beta=0.1
Gd=(matlab.tf([T,1],[alfa*T,1]))*alfa;Gd#位相遅れ
```

```python
Pdd=matlab.c2d(Gd, 0.001, method='zoh');Pdd
```

```python
matlab.bode(Pdd, np.logspace(-2, 2));
```
